##Task 1
Load the data and display the first 10 rows.

```python
# the data is from https://www.kaggle.com/datasets/dheerajmukati/most-runs-in-cricket?resource=download
path = '/content/sample_data/most_runs_in_cricket.csv'
df = pd.read_csv(path, skiprows=0)
df.head(10)
```

##Taks 2
Split the Player Column so that you have a sperate column for Player and Country. Also remove the brackets from the country column using .replace().

```python

df[['Player', 'Country']] = df['Player'].str.split('(', expand = True)
df['Country'] = df['Country'].str.replace(')',' ')
df.head(10)
```

##Task 3
Write a function to get the average of Total Matches played by player (Mat).

```python
def Avg_Mat(MAT):
  return np.nanmean(MAT)

print('The average MAT is: ' + str(Avg_Mat(df['Mat'])))
```

# Taks 4
Write a function to return the top 10 Total Matches played (Mat).

```python
def MAT(col):
 return col.nlargest(n=10)

print('Top 10 total matches played: ' + str(MAT(df['Mat'])) )
```

##Taks 5
Use Matplotlib to create a graph to display all the  player's  Names and total number of Not outs (No). Give your graph a title and remember to name the x and y axis. 

```python
import matplotlib.pyplot as plt

# Declaring the figure
plt.figure(figsize=[20, 16])

# Creating Horizontal plot
plt.barh(df['Player'],df['NO'], label = "TOP 10", color = 'g')

# Namimg the x and y axis
plt.xlabel('Total Not out')
plt.ylabel('Player Name')

# Giving the tilte for the plot
plt.title('Total Not Outs for each player')

# Displaying the bar plot
plt.show()

```
