```python
#Memo

#first import libraries
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
#load the CSV file
path = (r'C:\Users\Ramoraswi\Downloads\archive (1)\Age-standardized suicide rates.csv')
countries=pd.read_csv(path)
countries
```

```python
#average suicidal rates for each country across each year
Average_per_country = countries.groupby('Country').mean()
Average_per_country.reset_index(inplace = True)
Average_per_country
```

```python
#country with Highest suicidal rate in 2016
countries.sort_values(by=['2016']).head(1).Country
```

```python
#county with lowest suicidal rate in 2000
countries.sort_values(by=['2000']).tail(1).Country
```

```python
# plotting correlation heatmap
Heatmap = sns.heatmap(countries.corr())
plt.title('Heatmap that shows correlation between all the variables', fontsize=16)
# displaying heatmap
plt.show()
```

```python
#top 5 Countries with the most suicide rate in 2016
Top5Countries = pd.DataFrame()
Top5Countries = countries.sort_values(by=['2016'], ascending=False).head(5)
Top5Countries

```

```python
#5 countries with least suicide rate in the year 2016
Bottom5Countries = pd.DataFrame()
Bottom5Countries = countries.sort_values(by=['2016'], ascending=False).tail(5)
Bottom5Countries
```

```python
#ploting the graph
plt.figure(figsize=[14, 10])

# Creating Horizontal plot
plt.barh(Top5Countries['Country'], Top5Countries['2016'], label = "Top 5 countries ", color = 'r')
plt.barh(Bottom5Countries['Country'], Bottom5Countries['2016'], label = "Bottom 5 countries", color = 'g')

# Creating the legend of the bars in the plot
plt.legend()

# Namimg the x and y axis
plt.xlabel('Suicidal Rate')
plt.ylabel('Country')

# Giving the tilte for the plot
plt.title('TOP 5 vs Bottom 5 Countries based on their suicidal rate')

# Displaying the bar plot
plt.show()
```
